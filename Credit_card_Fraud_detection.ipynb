# Under Sampling
Building a sample dataset with similar distribution of normal transactions and Fraudulant transactions

No of Fraudulant Transactions --> 527

```python
legit_sample = legit.sample(n=527)
```

```python
new_dataset = pd.concat([legit_sample , fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset['V13'] = new_dataset['V13'].astype('float64')
```

```python
new_dataset.groupby('Class').mean()
```

### Splitting the data into Features and Targets.

```python
X= new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the data into Training and Testing Data.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_test.shape ,X_train.shape)
```

## Training the model & Evaluation by Accuracy score.

```python
# Training the model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Predicting test data
y_pred = model.predict(X_test)

# Calculating accuracy
test_accuracy = accuracy_score(Y_test, y_pred)
print("Logistic Regression Test Accuracy Score:", test_accuracy)

# Predicting training data
y_pred_train = model.predict(X_train)

# Calculating training accuracy
train_accuracy = accuracy_score(Y_train, y_pred_train)
print("Logistic Regression Training Accuracy Score:", train_accuracy)
```

Since both training and testing accuracies are high and relatively close to each other, therefore our model has no overfitting or underfitting and is a good fit.
